# Season Analysis

```python
match_per_season = df_ma.groupby('season')['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
sns.countplot(df_ma['season'])
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Season', fontsize=10)
plt.ylabel('count', fontsize=10)
plt.title('Total matches Played per season', fontsize=12, fontweight='bold')
```

```python
season_data = df_ma[['id', 'season']].merge(df_del, left_on='id', right_on='match_id', how='left').drop('id', axis=1)
season_data.head()
```

```python
season = season_data.groupby('season')['total_runs'].sum().reset_index()
p = season.set_index('season')
ax.set(facecolor='black')
ax = plt.axes()
sns.lineplot(data=p, palette='magma')
plt.title('Total runs in each season')
plt.show()
```

```python
runs_per_season = pd.concat([match_per_season, season.iloc[:,1]], axis=1)
runs_per_season['runs_per_match'] =  runs_per_season['total_runs']/runs_per_season['matches']
runs_per_season.set_index('season', inplace=True)
runs_per_season
```

```python
ax=plt.axes()
ax.set(facecolor='brown')
sns.barplot(data=runs_per_season.reset_index(), x='season', y='runs_per_match', palette='coolwarm')
plt.title('Average runs per Season')
plt.show()
```

```python
toss = df_ma['toss_winner'].value_counts()
ax = plt.axes()
ax.set(facecolor='grey')
sns.set(rc={'figure.figsize':(15,10)}, style='darkgrid')
ax.set_title('Number Of tosses won by each team', fontsize=15, fontweight='bold')
sns.barplot(y=toss.index, x=toss.values, orient='h', palette='icefire', saturation=1)
plt.xlabel('Number of Tosses won')
plt.ylabel('Team')
plt.show()
```

```python
ax = plt.axes()
ax.set(facecolor='grey')
sns.countplot(x='season', hue='toss_decision', data=df_ma, palette='magma', saturation=1)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('\n Season', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('Toss Decisions Across Seasons', fontsize=15, fontweight='bold')
plt.show()
```

```python
def col(df):
    if (df['win_by_runs'] == 0) and (df['win_by_wickets'] != 0):
        return 'wicket'
    elif (df['win_by_runs'] != 0 ) and (df['win_by_wickets'] == 0):
        return 'runs'
    else:
        return 'tie'
df_ma['win_type'] = df_ma.apply(col, axis=1)
```

```python
df_del['batsman_dismissed'] = np.where(df_del['batsman'] == df_del['player_dismissed'], 1,0)
```

```python
bat = df_del.groupby('batsman').agg({'batsman_runs':'sum', 'batsman_dismissed':'sum'}).reset_index()
bat.columns=['batsman', 'runs', 'times_dismissed']
bat.head()
```

```python
bat['batting_avg'] = np.where(bat['times_dismissed'] != 0, bat['runs']/bat['times_dismissed'], bat['runs'])
bat = bat.sort_values('batting_avg', ascending=False).head(15)
ax=plt.axes()
ax.set(facecolor='grey')
sns.barplot(x='batsman', y='batting_avg', data=bat, palette='rocket', saturation=1)
plt.xlabel('Batsman', fontsize=12)
plt.xticks(rotation=45)
plt.ylabel('Batting Average', fontsize=12)
plt.title('Top 15 Batsmen', fontsize=12, fontweight='bold')
plt.show()
```
